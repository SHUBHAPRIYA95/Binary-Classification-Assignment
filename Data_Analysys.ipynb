Balance the Dataset

```python
data['Class'].value_counts()
```

```python
non_class_dataset = data[data['Class']==0]
class_dataset = data[data['Class']==1]
```

```python
non_class_dataset.shape, class_dataset.shape
```

```python
non_class_dataset = non_class_dataset.sample(class_dataset.shape[0])
non_class_dataset.shape
```

```python
data = class_dataset.append(non_class_dataset, ignore_index=True)
data
```

```python
data['Class'].value_counts()
```

```python
sns.catplot(x='Class', data = data, kind = 'count')
```

Understanding the correlation between various features in the dataset



```python
correlation_matrix = data.corr()
```

constructing a heatmap to understand the correlation

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation_matrix, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Greens')
```

```python
for column in data:
    plt.figure()
    data.boxplot([column])
```
